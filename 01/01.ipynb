```python
q7 = data.loc[data['origin'] == 'Asia', ['vehicle_weight', 'model_year']]
q7_1 = q7.iloc[:7]
X = q7_1.values
XTX = np.dot(X.T, X)
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = np.dot(np.dot(XTX_inv, X.T), y)
sum(w)
```
