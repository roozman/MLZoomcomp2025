```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mutual_info_score
```

```python
data = pd.read_csv("course_lead_scoring.csv")
data.head()
```

```python
data.isna().sum()
```

```python
cols = data.columns
categorical = ['lead_source', 'industry',
       'employment_status', 'location']
numerical = ['number_of_courses_viewed', 'annual_income',
       'interaction_count', 'lead_score']
```

```python
data[categorical] = data[categorical].fillna("NA")
data[numerical] = data[numerical].fillna(0.0)
data.isna().sum()
```

```python
data['industry'].value_counts()
```

```python
data[numerical].corr().abs()
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(data, data['converted'], test_size = 0.2, random_state = 42) 
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, test_size=0.25, random_state=42)

del X_train['converted']
del X_valid['converted']
del X_test['converted']
```

```python
mi_scores = round(X_train[categorical].apply(lambda col: mutual_info_score(col, y_train)), 2)
mi_scores
# mutual_info_score(X_train['industry'], y_train)
```
